1. Difference between a function and a method in Python

ANS: A function is a block of code that performs a specific task and is independent of any object. A method is a function that is associated with an object and is called on that object.

Example:
# Function
def greet(name):
    return f"Hello, {name}!"

# Method
class Person:
    def greet(self, name):
        return f"Hello, {name}!"

2. Function arguments and parameters in Python

ANS: Parameters are the variables listed in the function definition. Arguments are the actual values passed to the function when it's called.

Example:
def add(a, b):  # a and b are parameters
    return a + b

add(3, 5)  # 3 and 5 are arguments

3.  Ways to define and call a function in Python
ANS: Functions can be defined using def keyword or as lambda functions. They can be called with positional arguments, keyword arguments, or default arguments.

Examples:
# Definition
def greet(name):
    return f"Hello, {name}!"

# Calling
greet("Alice")  # positional
greet(name="Alice")  # keyword

# Lambda function
square = lambda x: x*x
square(5)

4. Purpose of the return statement

ANS: The return statement ends the function execution and sends back the result to the caller. If no return statement, the function returns None.

Example:
def add(a, b):
    return a + b  # returns the sum

5. Iterators vs Iterables

ANS: An iterable is any object that can return an iterator (like lists, tuples). An iterator is an object that keeps state and produces the next value when next() is called.

Example:
my_list = [1, 2, 3]  # iterable
my_iter = iter(my_list)  # iterator
next(my_iter)  # returns 1

6. Generators in Python

ANS: Generators are functions that yield values one at a time using the yield keyword, maintaining their state between calls.

Example:
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

7. Advantages of generators

ANS: Memory efficient (produce items one at a time)
Can represent infinite sequences
Maintain state between calls
Lazy evaluation (only compute when needed)

8. Lambda functions

ANS: Lambda functions are small anonymous functions defined with the lambda keyword. They're typically used for short operations where a full function definition would be verbose.

Example:
square = lambda x: x*x

9. map() function

ANS: map() applies a given function to each item of an iterable and returns a map object (iterator).

Example:
numbers = [1, 2, 3]
squared = map(lambda x: x*x, numbers)

10. Difference between map(), reduce(), and filter()

ANS: map(): Applies function to all items in iterable

filter(): Filters items based on function condition

reduce(): Applies function cumulatively to reduce iterable to single value

Examples:
from functools import reduce

# map
list(map(lambda x: x*2, [1, 2, 3]))  # [2, 4, 6]

# filter
list(filter(lambda x: x > 2, [1, 2, 3, 4]))  # [3, 4]

# reduce
reduce(lambda x, y: x+y, [1, 2, 3, 4])  # 10

4. Prime number check (1-200)

ANS: def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def primes_up_to_200():
    return [n for n in range(1, 201) if is_prime(n)]

5. Fibonacci iterator

 ANS: class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1
        
    def __iter__(self):
        return self
        
    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

6. Powers of 2 generator

ANS: def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

7. File line reader generator

ANS: def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

10. Remove vowels with filter

ANS: def remove_vowels(s):
    vowels = 'aeiouAEIOU'
